```python
import os, json
from huggingface_hub import hf_hub_download
import rarfile
from datasets import Dataset
import pandas as pd
from PIL import Image

agri_repo = "Agri-LLaVA-Anonymous/Agricultural_pests_and_diseases_instruction_tuning_data"
data_dir = "./agri_llava_data"
image_dir = os.path.join(data_dir, "Img")
os.makedirs(data_dir, exist_ok=True)

def download_hf_file(filename, repo, local_dir):
    path = os.path.join(local_dir, filename)
    if not os.path.exists(path):
        print(f"Downloading {filename}...")
        hf_hub_download(repo_id=repo, filename=filename, repo_type="dataset", local_dir=local_dir)
    else:
        print(f"{filename} already exists.")
    return path

json_full_path = download_hf_file("agri_llava_instruction_tuning.json", agri_repo, data_dir)
json_1k_path   = download_hf_file("agri_llava_instruction_tuning_1k.json", agri_repo, data_dir)
rar_path       = download_hf_file("Img.rar", agri_repo, data_dir)

if not os.path.exists(image_dir):
    print(f"Extracting {rar_path}...")
    with rarfile.RarFile(rar_path) as rf:
        rf.extractall(data_dir)
    print("Extraction complete.")
else:
    print("Image directory already exists.")

with open(json_full_path, 'r') as f:
    data_full = json.load(f)
        
with open(json_1k_path, 'r') as f: 
    data_1k   = json.load(f)
df = pd.concat([pd.DataFrame(data_full), pd.DataFrame(data_1k)], ignore_index=True)
df.drop_duplicates(subset=['image'], inplace=True, keep='first')
raw_dataset = Dataset.from_pandas(df)

def sample_has_image(sample):
    if not sample.get('image'): return False
    return os.path.exists(os.path.join(image_dir, sample['image']))

image_dataset_raw = raw_dataset.filter(sample_has_image, num_proc=2)

TARGET = (512, 512)
system_message = "You are a helpful AI assistant specialized in crop disease diagnosis. Provide concise and accurate information."

def format_agri_data_for_trl(sample):
    image_path = os.path.join(image_dir, sample['image'])
    try:
        loaded_image = Image.open(image_path).convert("RGB").resize(TARGET, resample=Image.BILINEAR)
    except Exception:
        return None
    trl_messages = [{"role":"system","content":[{"type":"text","text":system_message}]}]  # FIXED: was tuple, now list
    for i, turn in enumerate(sample['conversations']):
        role = "user" if turn["from"] == "human" else "assistant"
        text_content = turn["value"]
        content_parts = []
        if role == "user" and i == 0 and loaded_image:
            content_parts.append({"type":"image", "image": loaded_image})
            text_content = text_content.replace("<image>", "").strip()
        if text_content:
            content_parts.append({"type":"text", "text": text_content})
        if not content_parts:
            content_parts.append({"type":"text","text":""})
        trl_messages.append({"role": role, "content": content_parts})
    return {"messages": trl_messages}

train_dataset = [format_agri_data_for_trl(s) for s in image_dataset_raw]
train_dataset = [x for x in train_dataset if x is not None]
print(f"Prepared messages for: {len(train_dataset)} samples")
```

```python
import matplotlib.pyplot as plt
from IPython.display import display, HTML

def create_image_gallery(dataset, num_samples=9):
    n = min(num_samples, len(dataset))
    cols = 3
    rows = (n + cols - 1) // cols
    fig, axes = plt.subplots(rows, cols, figsize=(4*cols, 4*rows))
    axes = axes.flat if n > 1 else [axes]
    i = 0
    for ax in axes:
        if i < n:
            msg = dataset[i]["messages"]
            img_obj = None
            for m in msg:
                for c in m["content"]:
                    if isinstance(c, dict) and c.get("type") == "image" and isinstance(c.get("image"), Image.Image):
                        img_obj = c["image"]
                        break
                if img_obj: break
            if img_obj:
                ax.imshow(img_obj)
                ax.set_title(f"Sample {i+1}")
            ax.axis("off")
            i += 1
        else:
            ax.axis("off")
    plt.suptitle("🌾 Agri-LLaVA Sample Images", fontsize=16, fontweight="bold")
    plt.tight_layout()
    plt.show()
create_image_gallery(train_dataset, num_samples=9)
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

img_obj = next(
    (c['image']
     for m in sample['messages']
     for c in m['content']
     if isinstance(c, dict) and c.get('type') == 'image'),
    None
)
if img_obj is None:
    raise ValueError("No image found in selected sample.")

display(img_obj.resize((512, 512)))

# Build widgets
question = widgets.Textarea(
    value="",
    description="❓ Question:",
    layout=widgets.Layout(width="100%", height="80px")
)
ask_btn = widgets.Button(description="🔬 Ask", button_style="success")
reset_btn = widgets.Button(description="🔄 Reset", button_style="info")
chat_out = widgets.Output()
chat_history = []

# Render the full conversation history
def render_chat():
    html = "<div class='section-card'><h3>💬 Conversation History</h3>"
    for turn in chat_history:
        html += (
            f"<div class='callout' style='margin:10px 0;'>"
            f"<strong>👤 You:</strong> {turn['user']}<br>"
            f"<strong>🤖 Assistant:</strong> "
            f"<pre style='white-space:pre-wrap;'>{turn['assistant']}</pre>"
            "</div>"
        )
    html += "</div>"
    chat_out.clear_output(wait=True)
    with chat_out:
        display(HTML(html))

# Call model with full history
def infer_with_history(image, history):
    messages = []
    for i, turn in enumerate(history):
        user_content = []
        if i == 0:
            user_content.append({"type": "image", "image": image})
        user_content.append({"type": "text", "text": turn["user"]})
        messages.append({"role": "user", "content": user_content})
        if "assistant" in turn:
            messages.append({
                "role": "assistant",
                "content": [{"type": "text", "text": turn["assistant"]}]
            })

    inputs = processor.apply_chat_template(messages, add_generation_prompt=True, tokenize=False)
    batch = processor(text=inputs, images=[[image]], return_tensors="pt")
    output = model.generate(**batch, max_new_tokens=256)
    text = processor.decode(output[0], skip_special_tokens=True)
    return text.split("model\n")[-1].strip()

# Button callback
def on_ask_click(_):
    q = question.value.strip()
    if not q:
        with chat_out:
            clear_output(wait=True)
            print("⚠️ Please enter a question.")
        return

    # append new turn with placeholder
    chat_history.append({"user": q, "assistant": "🧠 Generating response ..."})
    render_chat()

    # run inference and update the last assistant
    try:
        # model sees all past turns, but only the user text of last turn
        history_for_model = []
        for idx, turn in enumerate(chat_history):
            if idx == len(chat_history) - 1:
                history_for_model.append({"user": turn["user"]})
            else:
                history_for_model.append(turn)
        resp = infer_with_history(img_obj, history_for_model)
        chat_history[-1]["assistant"] = resp
        render_chat()
    except Exception as e:
        chat_history[-1]["assistant"] = f"Error: {e}"
        

def on_reset_click(_):
    chat_history.clear()
    chat_out.clear_output()

ask_btn.on_click(on_ask_click)
reset_btn.on_click(on_reset_click)

# display UI
display(widgets.VBox([question, widgets.HBox([ask_btn, reset_btn]), chat_out]))
```
